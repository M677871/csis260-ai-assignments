```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load the data
data = pd.read_csv("Salary_Data.csv")

# Explain the data
print("This dataset contains salary data vs years of experience.")
print("Each row represents one employee.")
print("We use:")
print("- YearsExperience as the input feature (X)")
print("- Salary as the target output (y)")
print("Shape:", data.shape)
print("Columns:", list(data.columns))
print(data.head())

# Keep only the two columns we need
data_small = data[["YearsExperience", "Salary"]].dropna()

# Correlation
print("Correlation:", data_small["YearsExperience"].corr(data_small["Salary"]))

X = np.c_[data_small["YearsExperience"]]
y = np.c_[data_small["Salary"]]

# Visualize the data (scatter)
data_small.plot(kind="scatter", x="YearsExperience", y="Salary")
plt.show()

# Histogram of salaries
plt.figure()
plt.hist(data_small["Salary"], bins=15, edgecolor="black")
plt.title("Histogram of Salaries")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.show()

# Select a linear model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Make a prediction
X_new = [[5]]  # 5 years of experience
print("Predicted salary for 5 years experience:", model.predict(X_new)[0][0])
```
