Project Report: Student Performance Prediction

1. Problem Description

The goal of this project is to examine whether study habits, specifically daily study time, are related to exam performance. We aim to explore the relationship between study_hours_per_day and exam_score and build a simple model that can predict an exam score from study hours.

2. Solution Implementation

Dataset Preparation:

We used the dataset student_habits_performance.csv and created a smaller dataset by randomly sampling 100 rows. The small dataset was saved as student_habits_performance_small.csv.

Data Explanation:

Each row represents one student record. We used study_hours_per_day as the input feature (X) and exam_score as the target output (y).

Feature Analysis:

We calculated the correlation between study_hours_per_day and exam_score and visualized the relationship using a scatter plot.

Modeling:

We trained a Linear Regression model (sklearn.linear_model.LinearRegression) on the sampled data to fit the relationship between X and y.

Results:

After training, we used the model to predict an exam score for a student who studies 5 hours per day using model.predict([[5]]).


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression


# Load the data
data = pd.read_csv("student_habits_performance.csv")

# Explain the data
print("This dataset contains student habits and their performance.")
print("Each row represents one student.")
print("We use:")
print("- study_hours_per_day as the input feature (X)")
print("- exam_score as the target output (y)")
print("Shape:", data.shape)
print("Columns:", list(data.columns))
print(data.head())

# Prepare the data (small dataset)
data_small = data.sample(n=100, random_state=42).reset_index(drop=True)
data_small.to_csv("student_habits_performance_small.csv", index=False)

# Use two columns based on correlation
print("Correlation:", data_small["study_hours_per_day"].corr(data_small["exam_score"]))

X = np.c_[data_small["study_hours_per_day"]]
y = np.c_[data_small["exam_score"]]

# Visualize the data
data_small.plot(kind='scatter', x="study_hours_per_day", y="exam_score")
plt.show()

# Select a linear model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Make a prediction
X_new = [[5]]  # 5 study hours per day
print("Predicted exam score for 5 study hours/day:", model.predict(X_new))
```
